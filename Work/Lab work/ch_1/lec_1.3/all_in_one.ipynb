## Q1
- Generate a normally distributed dataset using NumPy
- Plot Histogram and KDE
- Overlay Mean and Standard Deviation lines
- (Optional) Bell Curve using scipy

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm

data = np.random.normal(loc=50, scale=10, size=1000)

mean = np.mean(data)
std = np.std(data)

sns.histplot(data, kde=True)
plt.axvline(mean, color='#ff4c00', linestyle='--', label='Mean')
plt.axvline(mean + std, color='green', linestyle='--', label='+1 Std')
plt.axvline(mean - std, color='red', linestyle='--', label='-1 Std')

x = np.linspace(min(data), max(data), 100)
plt.plot(x, norm.pdf(x, mean, std) * len(data) * (max(data)-min(data))/50, label='Bell Curve')

plt.legend()
plt.title('Normal Distribution with Mean & Std Dev')
plt.show()
```

## Q2
- Load a dataset and calculate:
  - 25th percentile (Q1)
  - 50th percentile (Median)
  - 75th percentile (Q3)
- Plot quartiles using vertical lines

```python
import pandas as pd

df = pd.DataFrame({'Scores': data})

q1 = np.percentile(df['Scores'], 25)
q2 = np.percentile(df['Scores'], 50)
q3 = np.percentile(df['Scores'], 75)

sns.histplot(df['Scores'], kde=True)
plt.axvline(q1, color='red', label='Q1')
plt.axvline(q2, color='green', label='Median')
plt.axvline(q3, color='blue', label='Q3')
plt.legend()
plt.title('Quartiles Visualization')
plt.show()
```

## Q4
- Define Inferential Statistics
- List real-world use cases
- Compare population vs sample mean
- Visualize sample mean distribution

**Inferential Statistics** allows us to make predictions or decisions about a population based on sample data.

**Real-world use cases:**
- Election polling
- Clinical trials
- A/B testing
- Market research
- Quality control

```python
population = np.random.normal(100, 15, 5000)

population_mean = population.mean()
sample_means = []

for _ in range(200):
    sample = np.random.choice(population, size=100)
    sample_means.append(sample.mean())

sns.histplot(sample_means, kde=True)
plt.axvline(population_mean, color='red', label='Population Mean')
plt.legend()
plt.title('Sampling Distribution of Mean')
plt.show()
```
