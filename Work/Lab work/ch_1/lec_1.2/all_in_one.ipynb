Create a list to differentiate between them.

### Descriptive Statistics
- Summarizes data
- Uses measures like mean, median, mode
- Uses charts and tables

### Inferential Statistics
- Draws conclusions about a population
- Uses samples
- Includes hypothesis testing and confidence intervals

## Q2
- Load a sample dataset
- Use `describe()`
- Explain count, mean, std, min, max

```python
import pandas as pd

data = {
    'Scores': [55, 67, 72, 81, 90, 66, 74, 88, 92, 60]
}

df = pd.DataFrame(data)
df.describe()
```

## Q3
- Calculate Mean, Median, Mode using Pandas and NumPy
- Create a bar chart comparing Mean, Median, and Mode

```python
import numpy as np
import matplotlib.pyplot as plt

mean_val = df['Scores'].mean()
median_val = df['Scores'].median()
mode_val = df['Scores'].mode()[0]

mean_val, median_val, mode_val
```

```python
plt.bar(['Mean', 'Median', 'Mode'], [mean_val, median_val, mode_val])
plt.title('Mean vs Median vs Mode')
plt.ylabel('Score')
plt.show()
```

## Q4
- Calculate Range
- Calculate Variance (sample)
- Calculate Standard Deviation
- Create a boxplot and interpret dispersion

```python
range_val = df['Scores'].max() - df['Scores'].min()
variance_val = np.var(df['Scores'], ddof=1)
std_val = df['Scores'].std()

range_val, variance_val, std_val
```

```python
plt.boxplot(df['Scores'])
plt.title('Score Dispersion Boxplot')
plt.ylabel('Scores')
plt.show()
```
