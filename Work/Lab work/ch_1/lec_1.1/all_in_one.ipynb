## Q3. Pandas DataFrame Operations


```python
import pandas as pd

data = {
    'Name': ['Amit', 'Neha', 'Rahul', 'Sara'],
    'Age': [25, 30, 28, 35],
    'Department': ['Sales', 'HR', 'IT', 'Finance'],
    'Salary': [50000, 60000, 55000, 70000]
}

df = pd.DataFrame(data)
df
```

```python
df.dtypes
```

```python
df.select_dtypes(include='number')
```

## Q4. Data Visualization


```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure()
sns.histplot(df['Age'])
plt.title('Age Distribution')
plt.show()

sns.countplot(x='Department', data=df)
plt.title('Department Count')
plt.show()

df['Department'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Department Percentage')
plt.show()

sns.boxplot(y=df['Salary'])
plt.title('Salary Distribution')
plt.show()
```

## Q5. Population vs Sample


```python
import numpy as np

population = np.random.randint(0, 100, 1000)
sample = np.random.choice(population, 100)

population.mean(), population.std(), sample.mean(), sample.std()
```

**Explanation:** Sample statistics approximate population statistics but may differ due to randomness.
